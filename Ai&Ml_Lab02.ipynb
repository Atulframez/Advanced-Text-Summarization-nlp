Line Plot

Question:01

Write a Python program using Matplotlib to plot a line graph that shows the trend of values over time. Label the X-axis, Y-axis, and give a suitable title to the graph?

Theory

A line plot is used to display data points connected by straight lines. It is especially useful for showing trends or changes over time or over a sequence of values.

Explanation

The X-axis usually represents time or an ordered variable.

The Y-axis represents numerical values.

The line indicates whether the values are increasing, decreasing, or constant.

Line plots are widely used in sales analysis, temperature changes, and stock market trends.

```python
# 01) Line Plot (Trend over time)

import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y= [10,15,20,25,30]

plt.plot(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot Example")
plt.show()
```

Bar Chart

Question:02

Write a Python program using Matplotlib to create a bar chart for comparing values of different categories. Display appropriate axis labels and title?

Theory

A bar chart represents categorical data using rectangular bars. The length or height of each bar corresponds to the value of that category.

Explanation

Categories are placed on the X-axis.

Values are displayed on the Y-axis.

Bar charts make it easy to compare multiple categories at a glance.

Commonly used for performance comparison, survey results, and frequency analysis.

```python
# 02) Bar Chart (Category-wise comparison)

import matplotlib.pyplot as plt

categories = [ 'A', 'B', 'C', 'D']
values = [23, 45, 56, 12]

plt.bar(categories, values)
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Chart Example")
plt.show()
```

Histogram

Question:03

Write a Python program using Matplotlib to draw a histogram that represents the distribution of randomly generated data?

Theory

A histogram is used to show the distribution of numerical data by dividing the data into intervals known as bins.

Explanation

It helps identify how frequently values occur within specific ranges.

Shows whether data is normally distributed, skewed, or uniform.

Useful in statistical analysis, probability studies, and quality control.

```python
# 03) Histogram (Data distribution)
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(1000)

plt.hist(data, bins=30)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram Example")
plt.show()
```

Scatter Plot

Question:04

Write a Python program using Matplotlib to plot a scatter graph showing the relationship between two variables?

Theory

A scatter plot displays individual data points on a two-dimensional graph to show the relationship between two variables.

Explanation

Each point represents a pair of values.

Patterns in points indicate positive, negative, or no correlation.

Used in regression analysis, trend identification, and relationship studies.

```python
#4) Scatter Plot (Relationship between variables)
import matplotlib.pyplot as plt

x = [5,7,8,10,12]
y = [15,20,25,30,35]

plt.scatter(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot Example")
plt.show()
```

Box Plot

Question:05

Write a Python program using Matplotlib to create a box plot for identifying outliers in a given dataset?

Theory

A box plot summarizes data using quartiles and highlights outliers.

Explanation

The box represents the interquartile range.

The median shows the central value.

Whiskers show data spread.

Outliers indicate extreme or unusual values.

```python
#0 5) Box Plot (Outlier detection)

import matplotlib.pyplot as plt

data = [10, 12,14,15,18,19,100]

plt.boxplot(data)
plt.title("Box Plot Example")
plt.show()
```

Seaborn Scatter Plot

Question:06

Write a Python program using Seaborn to visualize the relationship between total bill and tip from a given dataset?

Theory

Seaborn builds on Matplotlib and provides enhanced statistical visualizations with better aesthetics.

Explanation

Scatter plots in Seaborn show relationships with improved clarity.

Often used in real-world datasets such as financial or business data.

Helps identify correlation and clustering patterns.

```python
# 06) Using Seaborn (Statistical Visualization)

import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.title("Seaborn Scatter Plot")
plt.show()
```

Heatmap

Question:07

Write a Python program using Seaborn to create a heatmap that displays the correlation matrix of a dataset?

Theory

A heatmap uses colors to represent values in a matrix and is commonly used to visualize correlations.

Explanation

Color intensity represents the strength of values.

Helps quickly identify strong and weak relationships.

Widely used in machine learning and data analysis.

```python
# 07) Heatmap (Correlation matrix)

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data = pd.DataFrame(np.random.rand(5,5), columns=list("ABCDE"))
sns.heatmap(data , annot= True)
plt.title("Heatmap Example")
plt.show()
```
