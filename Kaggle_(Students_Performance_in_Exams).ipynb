## Q3. Average Scores Grouped by Gender Theory

Grouping data allows comparison between categories. Calculating average scores by gender helps identify performance differences.

Explanation

Scores are grouped into male and female categories.

Mean values of Math, Reading, and Writing scores are calculated.

This analysis highlights subject-wise gender performance trends.

```python
# Bar graph
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("StudentsPerformance.csv")

# Calculate average scores by gender
avg_scores = df.groupby("gender")[["math score", "reading score", "writing score"]].mean()
avg_scores = avg_scores.reset_index().melt(
    id_vars="gender",
    var_name="Subject",
    value_name="Average Score"
)

# Bar chart
plt.figure(figsize=(8, 5))
sns.barplot(x="gender", y="Average Score", hue="Subject", data=avg_scores)
plt.title("Average Exam Scores by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Score")
plt.show()
```

```python
avg_scores = df.groupby("gender")[["math score","reading score","writing score"]].mean()

avg_scores.plot(kind="bar", figsize=(8,6))
plt.title("Average Scores by Gender")
plt.ylabel("Average Score")
plt.xticks(rotation=0)
plt.show()
```

## Q5. Scatter Plot between Reading and Writing Scores Theory

Scatter plots visualize the relationship between two numerical variables.

Explanation

Each point represents a student.

Points forming a diagonal pattern indicate strong correlation.

Shows how reading performance relates to writing ability.

```python
# Scatter Plot
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("StudentsPerformance.csv")

# Scatter plot
plt.figure(figsize=(8, 5))
sns.scatterplot(
    x="reading score",
    y="writing score",
    data=df,
    alpha=0.6
)
plt.title("Scatter Plot of Reading vs Writing Scores")
plt.xlabel("Reading Score")
plt.ylabel("Writing Score")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(
    x="reading score",
    y="writing score",
    data=df,
    hue="gender"
)
plt.title("Reading vs Writing Scores")
plt.show()
```

## Q 06. Heatmap of Correlation Matrix Theory

A heatmap represents correlation values using colors.

Explanation

Darker colors represent stronger relationships.

Makes correlation interpretation easier and faster.

Confirms strong interdependence between subjects.

```python
# Heatmap
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("StudentsPerformance.csv")

# Correlation matrix
corr = df[["math score", "reading score", "writing score"]].corr()

# Heatmap
plt.figure(figsize=(7, 5))
sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap of Exam Scores")
plt.show()
```

```python
corr = df[["math score","reading score","writing score"]].corr()

plt.figure(figsize=(6,4))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
corr = df[["math score","reading score","writing score"]].corr()

plt.figure(figsize=(16,8))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

## Q 07. Correlation Heatmap for Standard Lunch Students Theory

External factors such as nutrition may affect academic performance.

Explanation

Data is filtered for students with standard lunch.

Correlation heatmap shows stronger score relationships.

Indicates better consistency in performance for students with proper nutrition.

```python
standard_lunch_corr = df[df["lunch"] == "standard"][
    ["math score","reading score","writing score"]
].corr()

plt.figure(figsize=(6,4))
sns.heatmap(standard_lunch_corr, annot=True, cmap="cubehelix")
plt.title("Correlation (Standard Lunch Students)")
plt.show()
```

## Q 08. Correlation Heatmap for Female Students Theory

Analyzing subsets helps identify group-specific patterns.

Explanation

Heatmap focuses only on female students.

Correlations are compared with overall data.

Helps understand gender-based academic trends.

```python
female_corr = df[df["gender"] == "female"][
    ["math score","reading score","writing score"]
].corr()

plt.figure(figsize=(6,4))
sns.heatmap(female_corr, annot=True, cmap="magma")
plt.title("Correlation Heatmap (Female Students)")
plt.show()
```
